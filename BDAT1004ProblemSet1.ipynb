```python
import math
a = int(input('Enter the value for 1st side of triangle: '))
b = int(input('Enter the value for 2nd side of triangle: '))
c = int(input('Enter the value for 3rd side of triangle: '))

def triangleArea(a,b,c):
    s = (a+b+c)/2
    area = math.sqrt(s*(s-a)*(s-b)*(s-c))
    print('Area of the triangle: ', area)
    
triangleArea(a,b,c)
```

# Question 5 C#

```python
def inside(x,y,x1,y1,x2,y2):
    
    if (x > x1 and x < x2) and (y > y1 and y < y2):
        print("True")
    else:
        print("False")
```

```python
inside(1,1,0,0,2,3)
inside(-1,-1,0,0,2,3)
inside(1,1,0.3,0.5,1.1,0.7)
inside(1,1,0.5,0.2,1.1,2)
```

```python
string = input("Enter the string: ")
def pig(string):
    string = str.lower(string);
    vowel = "aeiou"
    
    if (string[0] in vowel):
        string = string + "way"
        print(string)
        
    else:
        dummy = string[0]
        string = string[1:]
        string = string + dummy + "ay"
        print(string)
pig(string)
```

# Question 7 Python

```python
def bldcount(text):
    file = open(text, "r")
    list = file.readlines() 
    for i in list:
        a = i.split() 

    A = a.count('A')
    B = a.count('B')
    AB = a.count('AB')
    O = a.count('O')
    OO = a.count('OO')
    print("There are" ,A, "patients with blood group A")
    print("There are" ,B, "patients with blood group B")
    print("There are" ,AB, "patients with blood group AB")
    print("There are" ,O, "patients with blood group O")
    print("There are" ,OO, "patients with blood group OO")
    
bldcount('bloodtype1.txt');
```

# Question 8 Python

```python
def curconv(text, rate):
    currencies = open("currencies.txt","r");
    i = {};
    for cur in currencies:
        i[cur[:3].strip()] = cur[4:].split("\t");
    convRate = i[text][0].strip();
    print(rate * float(convRate));

curconv('EUR', 100);
```

# Question 10 Python

```python
def frequencies(string):   
    alphabets=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    cList = []; 
    output = {};
    for word in string:
        for char in word:
            if char.isalpha():
                cList.append(char);         
    for a in alphabets:
        if a in cList:
            output[a] = cList.count(a);
        else:
            output[a] = 0;

    fList = (output.values());
    print(fList);

frequencies('apple');
```
